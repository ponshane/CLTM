```python
# JSD Function
def calculate_jsd_of_theta_from_MTAnchor(theta1, theta2):
    assert theta1.shape[0] == theta2.shape[0]
    
    n_docs = theta1.shape[0]
    jsd_array = []
    
    for idx in range(n_docs):
        try:
            jsd_array.append(jsd(theta1[idx], theta2[idx]))
        except TypeError:
            print("array does not meet the requirement!")
            break

    return jsd_array
```
